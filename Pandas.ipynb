### 1.what the region and sex wise avg insurance charges?

```python


round(insurance.groupby(by=['sex','region'])['charges'].mean())
```

```python

round(pd.pivot_table(data=insurance,values='charges',index='region',columns='sex',aggfunc='mean'))
```

```python

pd.crosstab(index = insurance['region'], columns = insurance['sex'], margins=True) #simple cross tabulation of two (or more) factors. And that factors need to be categorical.
```

```python
round(insurance.groupby(by=['sex','region'])['age'].mean())
```

```python
round(pd.pivot_table(data=insurance,values='age',index='region',columns='sex',aggfunc='mean'))
```

### Add the age by 2

```python
student_details['age'] = student_details['age'].apply(lambda x: x-6)
student_details
```

```python
student_details['Fedu'] = student_details['Fedu'].apply(lambda x: x*2)
student_details
```

```python
student_details['Fjob'] = student_details['Fjob'].apply(lambda x:x.upper())
student_details
```
