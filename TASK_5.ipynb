### 1.Merge
* The merge() method updates the content of two DataFrame by merging them together, using the specified method(s).

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Name': ['Karunas', 'Senthil', 'Kamesh', 'Shyam'],
                    'Dept': ['EEE', 'Auto', 'E&I', 'Civil']})
df2 = pd.DataFrame({'Name': ['Senthil', 'Shyam', 'Karunas', 'Kamesh'],
                    'Percentage': [75, 80, 65, 73]})
df1
```

```python
df2
```

### one to one joints

```python

df3=pd.merge(df1,df2)
df3
```

### many to one joins

```python

df4 = pd.DataFrame({'Dept':['Auto','EEE','E&I','Civil'],
                    'Location':['C-block','B-block','C-block','D-block']})
```

```python
df5 = pd.merge(df3,df4)
df5
```

### many to many joins

```python

df6 = pd.DataFrame({'Dept':['EEE','Auto','E&I','Civil','Auto','EEE','EEE','EEE'],
                    'Skills':['CAD Design','3D Design','Instrumentation','ED','Bike','EMD','Power system','iEEE']})
df7 = pd.merge(df6,df5)
df7
```

### using keyword

```python
a = pd.merge(df1,df2,on='Name')
a
```

```python
stipend = pd.DataFrame({'Names': ['Senthil', 'Shyam', 'Karunas', 'Kamesh'],
                    'stipend': [15000, 18000, 16000, 17000]})
```

```python
b=pd.merge(df1, stipend, left_on="Name", right_on="Names")
b
```

```python
b.drop(labels='Names',axis=1)
```

### join

```python
df6 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],
                    'food': ['fish', 'beans', 'bread']},
                   columns=['name', 'food'])
df7 = pd.DataFrame({'name': ['Mary', 'Joseph'],
                    'drink': ['wine', 'beer']},
                   columns=['name', 'drink'])
df6
```

```python
df7
```

### inner join

```python
k=pd.merge(df6,df7,how = 'inner')
k
```

### outer join

```python
m=pd.merge(df6,df7,how = 'outer')
m
```

### left join


```python
n=pd.merge(df6,df7,how = 'left')
n
```

### right join

```python
o =pd.merge(df6,df7,how = 'right')
o
```

### 3.1 Lambda(Use apply function with lambda)
* A lambda function is a small anonymous function.

* A lambda function can take any number of arguments, but can only have one expression.

```python
student = pd.read_csv('student-mat.csv')
student.head()#student-mat.csv
```

### Eg-1

```python
student['age'] = student['age'].apply(lambda a:a+2)
student.head()
```

### Eg-2

```python
student['health'] = student['health'].apply(lambda o:o+2)
student.head()
```

### 3.2.user defined function(Use apply with user-defined function)

### Eg-1

```python
def address(a):
    if a == 'U':
        return 'Easy'
    elif a == 'R':
        return 'Moderate'
    else:
        return 'Difficult'
    
student['Transport'] = student['address'].apply(address)  
student
```

```python
student['Transport'].value_counts()
```

### Eg-2

```python
def absence(f):
    if f == 0:
        return "BEST"
    elif f>4:
        return"GOOD"
    else:
        return"BAD"
student['Attendnace']= student['absences'].apply(absence)
student
```

```python
student['Attendnace'].value_counts()
```

```python
### analysis
round(pd.pivot_table(student,values='absences',index='sex',columns='address'))#,columns='Transport')
```

### From the above analysis, Absences in Rural area is greater than Urban area students especially Female students
