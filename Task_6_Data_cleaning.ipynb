```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
sales_2017=pd.read_csv('Sales Transactions-2017.csv')
sales_2017
```

```python
sales_2018=pd.read_csv('Sales Transactions-2018.csv')
sales_2018
```

```python
sales_2019=pd.read_csv('Sales Transactions-2019.csv')
sales_2019
```

```python
sales_data = pd.concat([sales_2017,sales_2018,sales_2019])
sales_data
```

```python
sales_data.dtypes
```

```python
sales_data.isnull().sum()
```

## 1.droping unwanted columns

```python
sales_data.drop(['Gross','Disc','Voucher Amount'],axis=1,inplace=True)
sales_data
```

```python
sales_data.dropna(axis=0,inplace=True)
sales_data
```

```python
sales_data.head(20)
```

```python
sales_data.replace(" ", np.nan, inplace=True)
sales_data
```

```python
sales_data.head(20)
```

```python
sales_data.dropna(axis=0,subset=['Date','Voucher','Party','Product','Qty','Rate'],inplace=True)
sales_data
```

```python
sales_data.head(31)
```

```python
sales_data.reset_index(drop=True,inplace=True)
sales_data.head(20)
```

```python
sales_data.dtypes
```

## 2.convert the data types
#### 1.Date(object into datetime)

```python
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
sales_data.dtypes
```

#### 2.Vocuher

```python
sales_data['Voucher']=sales_data['Voucher'].apply(lambda a:a.replace('Sal:',' ')).astype(int)
sales_data
```

### 3.Party,Product,rate

```python
sales_data['Party']=sales_data['Party'].str.upper()
sales_data
```

```python
sales_data['Product']=sales_data['Product'].str.upper()
sales_data
```

```python
sales_data['Qty'] = sales_data['Qty'].str.replace(',','').astype(float)
sales_data['Rate'] = sales_data['Rate'].str.replace(',','').astype(float)
sales_data
```

```python
sales_data['Qty'] = sales_data['Qty'].astype(int)
sales_data.dtypes
```

```python
sales_data.isnull().sum()
```
